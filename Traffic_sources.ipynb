### Задача - определить источник трафика

Создаем столбец source_type по следующим правилам:

- если источник traffic_source -- yandex или google, то в source_type ставится organic
- для источников paid и email из России ставим ad
- для источников paid и email не из России ставим other
- все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
df1 = pd.read_csv('visit_log.csv', ';')
df1.head()
```

```python
def source (df):
    if df['traffic_source'] == ('yandex' or 'google'):
        return 'organic'
    elif df['region'] != 'Russia' and df['traffic_source'] == ('paid' or 'email'):
            return 'other'
    elif df['region'] == 'Russia' and df['traffic_source'] == ('paid' or 'email'):
            return 'ad'
    else:
        return df['traffic_source'] 
    
```

```python
df1['source_type'] = df1.apply(source, axis = 1)
df1.tail()
```
